```python
from plotly import __version__
from plotly.offline import init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import pandas as pd
import sqlite3 as sql


init_notebook_mode(connected=True)
```

```python
conn = sql.connect('crypto_signal/db.sqlite3')
```

```python
res = conn.execute("SELECT name FROM sqlite_master WHERE type='table';")
for name in res:
    print (name[0])
```

```python
cursor = conn.cursor()
```

```python
df = pd.read_sql(con=conn, sql="select * from crypto_track_pytrends")
df.tail()
df['date'] = pd.to_datetime(df['date'].str.strip(), format='%Y/%m/%d')
df.dtypes
```

```python
trace0 = go.Scatter(
    x = df['date'],
    y = df['trend_ratio'],
    mode = 'lines',
    name = 'lines'
)
data =[trace0]
iplot(data)
```
